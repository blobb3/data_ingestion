### Aufgabe:
Öffnen Sie die URL https://en.wikipedia.org/wiki/Web_scraping und verwenden Sie XPath, um den ersten Absatz (<p>) im Hauptinhalt der Seite zu finden und anzuzeigen.

- Frage: Wie unterscheidet sich der XPath für verschachtelte Elemente von einfacheren Selektoren?
- Hinweis: Der XPath könnte z. B. //*[@id="mw-content-text"]/div/p[1] sein.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

# Setup für Selenium WebDriver im Headless-Modus
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Pfad zum ChromeDriver angeben
service = Service('/usr/local/bin/chromedriver')  # Passen Sie den Pfad bei Bedarf an
driver = webdriver.Chrome(service=service, options=options)

# URL öffnen
driver.get("https://en.wikipedia.org/wiki/Web_scraping")

# Warten, um sicherzustellen, dass die Seite vollständig geladen ist
time.sleep(2)

# Finden des ersten Absatzes im Hauptinhalt mit verschachteltem XPath
paragraph = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/p[1]').text
print("Erster Absatz im Hauptinhalt der Seite:")
print(paragraph)

# Browser schließen
driver.quit()
```

### Aufgabe 2: Screenshot einer bestimmten Website

##### Aufgabe:
Öffnen Sie die DinoWebseite mit Selenium und machen Sie einen Screenshot der gesamten Seite. Speichern Sie das Bild als dino_screenshot.png.

- Frage: Warum ist der Headless-Modus bei dieser Art von Aufgabe nützlich?
- Antwort: 
    1. Ressourcenschonend und schneller: Im Headless-Modus wird die grafische Benutzeroberfläche des Browsers nicht geladen. Dadurch werden weniger Systemressourcen wie CPU und RAM verwendet, was die Geschwindigkeit der Tests oder des Web-Scrapings verbessert.
    2. Für automatisierte Umgebungen (z. B. Server, Cloud) geeignet
    3. Bessere Stabilität und Automatisierung


```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import time

# Setup für Selenium WebDriver im Headless-Modus
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Verwenden von WebDriver Manager zur automatischen Installation des ChromeDrivers
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# DJI Store-Seite öffnen und warten, bis sie geladen ist
driver.get('https://www.dinoworld.ch/')
time.sleep(3)  # Warte 3 Sekunden

# Screenshot erstellen
driver.save_screenshot('dino_screenshot.png')

# Screenshot anzeigen
img = mpimg.imread('dino_screenshot.png')
plt.figure(figsize=(8, 8))
plt.axis('off')
plt.imshow(img)
plt.show()

# Browser schließen
driver.quit()
```

### Aufgabe 3: Verwenden von XPath zur Extraktion von verschachtelten Elementen

#### Aufgabe:
Öffnen Sie die Seite „Hot Rod“ auf Wikipedia (https://de.wikipedia.org/wiki/Hot_Rod) und verwenden Sie XPath, um die Bildunterschrift (Caption) des ersten Bildes im Hauptinhalt der Seite abzurufen.

- XPath-Hinweis: Versuchen Sie, //*[@id="mw-content-text"]/div[1]/figure[1]/figcaption zu verwenden.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

# Setup für Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service, options=options)

# Wikipedia-Seite öffnen
driver.get('https://de.wikipedia.org/wiki/Hot_Rod')

# Finden der Bildunterschrift mit XPath
caption = driver.find_element(By.XPATH, '/html/body/div[3]/div[3]/div[5]/div[1]/figure[1]/figcaption').text
print("Bildunterschrift des ersten Bildes:", caption)

# Browser schließen
driver.quit()
```

#### Aufgabe 4: Automatische Suche und Navigation auf Wikipedia

Aufgabe:
Öffnen Sie die Hauptseite von Wikipedia in Deutsch (https://de.wikipedia.org), geben Sie das Suchwort „Python (Programmiersprache)“ ein und führen Sie die Suche aus. Zeigen Sie die Einleitung des Artikels im Konsolenoutput an.

- XPath für das Suchfeld: //*[@id="searchInput"]

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

# Setup für Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service, options=options)

# Wikipedia-Hauptseite öffnen
driver.get('https://de.wikipedia.org')

# Suchfeld finden und Text eingeben
search_field = driver.find_element(By.XPATH, '//*[@id="searchInput"]')
search_field.send_keys('Käseherstellung')

# Suche ausführen
search_field.submit()

# Kurze Wartezeit, damit die Seite geladen wird
time.sleep(2)

# Einleitungstext im Artikel extrahieren
intro_text = driver.find_element(By.XPATH, '/html/body/div[3]/div[3]/div[5]/div[1]/p[2]').text
print("Einleitung des Artikels:", intro_text)

# Browser schließen
driver.quit()
```

### Alle H3 titel einer Webseite ausgeben
Ziel der Aufgabe: Verwenden Sie Selenium, um alle h3-Titel auf einer bestimmten Webseite zu finden und auszugeben.
Aufgabenstellung:

- Öffnen Sie die Webseite https://example.com (Sie können eine beliebige Webseite wählen, die h3-Titel enthält).
- Verwenden Sie Selenium und die Methode find_elements, um alle h3-Titel auf der Seite zu finden.
- Geben Sie jeden gefundenen h3-Titel in der Konsole aus.

Frage:

- Welche Methode von Selenium wird verwendet, um mehrere Elemente einer bestimmten Tag-Art auf einer Seite zu finden?
- Erklären Sie den Unterschied zwischen find_element und find_elements.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Optional: headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Start the Chrome driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Website zum Scrapen öffnen
driver.get('https://blog.dataiku.com/')  # Ersetzen Sie diese URL durch eine Seite mit <h3>-Titeln

# Kurze Wartezeit, um sicherzustellen, dass die Seite vollständig geladen ist
time.sleep(2)

# Alle <h3>-Elemente auf der Seite finden
h3_elements = driver.find_elements(By.TAG_NAME, 'h3')

# Die Texte aller gefundenen <h3>-Elemente ausgeben
print("Alle <h3>-Titel auf der Seite:")
for element in h3_elements:
    print(element.text)

# Driver schließen
driver.close()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Optional: headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Start the Chrome driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Website zum Scrapen öffnen
driver.get('https://blog.dataiku.com/')  # Ersetzen Sie diese URL durch eine Seite mit spezifischen id- und class-Attributen

# Kurze Wartezeit, um sicherzustellen, dass die Seite vollständig geladen ist
time.sleep(2)

# 1. Suche nach einem Element anhand der ID
try:
    main_content = driver.find_element(By.ID, 'main-content')
    print("Inhalt des Elements mit ID 'main-content':")
    print(main_content.text)
except:
    print("Element mit ID 'main-content' nicht gefunden.")

# 2. Suche nach mehreren Elementen anhand des Klassennamens
section_titles = driver.find_elements(By.CLASS_NAME, 'section-title')

print("\nInhalte aller Elemente mit Klassenname 'section-title':")
for element in section_titles:
    print(element.text)

# Driver schließen
driver.close()
```
